```python
####changing  years in current job from object to float

data['Years_in_current_job'] = data['Years_in_current_job'].str.replace(' years', '', regex=False)
data['Years_in_current_job'] = data['Years_in_current_job'].str.replace('< 1 year', '0.5', regex=False)
data['Years_in_current_job'] = data['Years_in_current_job'].str.replace('10+', '10', regex=False)

data['Years_in_current_job'] = data['Years_in_current_job'].astype(float)

data.head()
```

```python
##imputing for missing values for years in current job
median_years_in_job = data['Years_in_current_job'].median()
data['Years_in_current_job'] = data['Years_in_current_job'].fillna(median_years_in_job)

# Verify the imputation
print(data['Years_in_current_job'].isnull().sum())

```

```python
#imputing missing values for months since last delinquency
median_Months_since_last_delinquency = data['Months_since_last_delinquency'].median()
data['Months_since_last_delinquency'] = data['Months_since_last_delinquency'].fillna(median_Months_since_last_delinquency)
print(data['Months_since_last_delinquency'].isnull().sum())
```

```python
##imputing missing value for bankruptciences
median_bankruptcies = data['Bankruptcies'].median()
data['Bankruptcies'] = data['Bankruptcies'].fillna(median_bankruptcies)
print(data['Bankruptcies'].isnull().sum())
```

```python
##imputing missing data for taxliens
median_Tax_Liens = data['Tax_Liens'].median()
data['Tax_Liens'] = data['Tax_Liens'].fillna(median_Tax_Liens)
print(data['Tax_Liens'].isnull().sum())
```

```python
purpose_counts = datacleaned['Purpose'].value_counts()

# Get labels and sizes from the value counts
labels = purpose_counts.index
sizes = purpose_counts.values

# Plot
plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Purpose Distribution')
plt.axis('equal')  # Makes it a circle
plt.show()
```

```python
#correlation matrix
correlation_matrix = datacleaned.corr(numeric_only=True)
correlation_matrix
```

```python
num_features = len(correlation_matrix.columns)

# The width and height scale with the number of features
fig_width = num_features * 1.2   # adjust width factor
fig_height = num_features * 0.7  # adjust height factor for row size

plt.figure(figsize=(fig_width, fig_height))

# Create heatmap
sns.heatmap(correlation_matrix,
            annot=True,        # show correlation coefficients
            fmt=".4f",         # 4 decimal places
            cmap='coolwarm',   # color palette
            linewidths=0.5,    # lines between cells
            cbar=True,         # show color bar
            square=True)       # keep squares

# Rotate axis labels for readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Add a title
plt.title('Correlation Heatmap')

# Adjust layout to prevent clipping
plt.tight_layout()

# Show plot
plt.show()
```


**Interpretation**: The Heatmap above shows that most of the off-diagonal values are very close to 0.00 (e.g., 0.0072, -0.0047, 0.0035), which indicates that there is very little linear correlation between the numeric features. For example, the correlation between Credit_Score and Current_Loan_Amount is about -0.0042, showing no relationship. Overall, this shows low multicollinearity in the dataset, which is beneficial for machine learning model, as it reduces the risk of redundant predictors causing instability or inflated variance in model estimates.
