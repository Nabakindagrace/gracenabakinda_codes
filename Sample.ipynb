```python
def find_most_frequent_char(s):
    """
    Finds the character that appears the most times in a given string.

    Args:
        s (str): The input string.

    Returns:
        str: The most frequent character, or an empty string if the input is empty.
    """

    char_counts = {}
    for char in s:
        char_counts[char] = char_counts.get(char, 0) + 1
    if not char_counts:
        return ""

    max_count = 0
    most_frequent_char = ''

    for char, count in char_counts.items():
        if count > max_count:
            max_count = count
            most_frequent_char = char

    return most_frequent_char

# --- Example Usage ---
string1 = "programming"
string2 = "teacher"
string3 = "google"

print(f"The most frequent character in '{string1}' is: '{find_most_frequent_char(string1)}'")
print(f"The most frequent character in '{string2}' is: '{find_most_frequent_char(string2)}'")
print(f"The most frequent character in '{string3}' is: '{find_most_frequent_char(string3)}'")
```

```python
def count_vowels(s):
    """
    Counts the total number of vowels (a, e, i, o, u) in a given string,
    regardless of case.

    Args:
        s (str): The input string.

    Returns:
        int: The total count of vowels.
    """

    VOWELS = {'a', 'e', 'i', 'o', 'u'}

    vowel_count = 0

    s_lower = s.lower()

    for char in s_lower:

        if char in VOWELS:
            vowel_count += 1

    return vowel_count

# --- Example Usage ---
string1 = "hello"
string2 = "ProgrammingIsFun"
string3 = "AEIOU"

print(f"The number of vowels in '{string1}' is: {count_vowels(string1)} (e, o)")
print(f"The number of vowels in '{string2}' is: {count_vowels(string2)} (o, a, i, I, u)")
print(f"The number of vowels in '{string3}' is: {count_vowels(string3)} (A, E, I, O, U)")
```
